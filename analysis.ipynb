```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot_history(set: str, setting: str):
    """Plot the history of the training of the model

    Args:
        file (str): Path to the history file
    """
    train_history = pd.read_csv(f"evaluation/{set}/history/train_{setting}.csv")
    test_history = pd.read_csv(f"evaluation/{set}/history/test_{setting}.csv")
    sns.lineplot(data=train_history, x="time", y="precision", label="Precision Train")
    sns.lineplot(data=test_history, x="time", y="precision", label="Precision Test")
    sns.lineplot(data=train_history, x="time", y="recall", label="Recall Train")
    sns.lineplot(data=test_history, x="time", y="recall", label="Recall Test")
    plt.ylim(-0.05, 1.05)
    plt.xlim(0,300)
    plt.title(f"Training and Test Score for {set} with {setting}")
    plt.show()
```

```python
plot_history("supports","popper_all_dl_bkcons")
plot_history("supports","popper_all_dl")
plot_history("supports","popper_all")
plot_history("stable","popper_all_dl_bkcons")
plot_history("stable","popper_all_dl")
plot_history("stable","popper_all")
```

```python
def plot_eval(set: str, setting: str):
    """Plot the evaluation of the models

    Args:
        file (str): Path to the history file
    """
    eval = pd.read_csv(f"evaluation/{set}/{setting}.csv")
    # Sort the dataframe by the solution
    eval = eval.sort_values(by=["solution"])
    # Rotate the labels on the x-axis
    plt.xticks(rotation=90)
    sns.barplot(data=eval, x="solution", y="f1", label="F1")
    plt.ylim(0, 1)
    plt.title(f"Evaluation Score for {set}")
    plt.show()

plot_eval("supports", "all")
plot_eval("stable", "all")
```
